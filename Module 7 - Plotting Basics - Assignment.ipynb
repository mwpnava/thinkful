```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline


names = ['George','John','Thomas','James','Andrew','Martin','William','Zachary','Millard','Franklin']

df = pd.DataFrame(index=names)

df['Names'] = names
df['Country'] = ['US','CAN','CAN','US','CAN','US','US','US','CAN','US']

df['ad_views'] = [16,42,32,13,63,19,65,23,16,77]

df['items_purchased'] = [2,1,0,8,0,5,7,3,0,5]
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['ad_views'], color='red')
plt.title('Ad Views')


plt.subplot(1,2,2)
plt.plot(df['items_purchased'])
plt.title('Items Purchased')

plt.tight_layout()
plt.show()
```

Graphics above are ad views and items purchased by 10 customers

```python
##Creating a simple plot

plt.plot(df['ad_views'], color='purple')
plt.plot(df['items_purchased'], color='green')
plt.ylabel('Values')
plt.title('Ad Views vs Items Purchased')
```

The graphic above depicts Ad Views and Items Purchases, it seems that is not a direct relationship between ad views and items purchased. It means that is not correct to say the more a customer watchs ads the more that customer purchases items 

```python
##Let's calculate the penetration percentage of ad views based on the purchased items.

df['penetration_percentage'] = round((df['items_purchased'] / df['ad_views']) * 100,2)
df
```

```python
plt.scatter(x=df['Country'], y=df['penetration_percentage'], c=['yellow','green'])
plt.title('Penetration percentage of ad views per country')
plt.show()
```

Graphic above shows that penetration percentage of ad views is higher in US than CA
