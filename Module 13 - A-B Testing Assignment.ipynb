### Does a new supplement help people sleep better?

A) 2 versions: 

1. Taking a new supplement
2. Taking a another supplement

B) Samples

1. Group of people  who takes new suplement
2. Group of people who takes regular suplemens
3. Group of people who takes both suplements (every other day)

Both groups are similar in age and kind of jobs

C) Hypothesis:

People who takes new suplement sleeps better than people who don't

D) Outcomes:  

- Number of continuous hours during sleeping time
- Number of interruptions during sleeping time
- Feeling measurement at the morning

E) Other Measurements:

- Consider how much light is int the place to sleep
- Consider how much noise is in the place to sleep
- Consider the previous activity before go to sleep
