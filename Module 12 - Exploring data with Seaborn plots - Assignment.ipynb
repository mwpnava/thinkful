### 1. Find a dataset you'd like to explore. This can be something you're familiar with or something new. Create a Jupyter notebook and then:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set()

# Reading from a csv file and creating the dataset 
goodreads = pd.read_csv('good_reads_final.csv')
goodreads.head(5)
```

```python
# Chosing a variable and plot that variable four different ways

author_avg_r = goodreads['author_average_rating']
sns.relplot(x="author_average_rating", y="author_rating_count", data=goodreads);
sns.relplot(x="author_average_rating", y="author_rating_count", kind="line", data=goodreads);
sns.catplot(x="author_average_rating",y="author_rating_count", data=goodreads);
```

```python
# Choose two continuous variables, and plot them three different ways.

#author_average_rating
#num_reviews

sns.relplot(x="num_reviews", y="num_ratings", kind="line", data=goodreads)
sns.relplot(x="num_reviews", y="num_ratings", data=goodreads);
sns.relplot(x="num_reviews", y="num_ratings", hue="author_gender", data=goodreads);
```

```python
# Choose one continuous variable and one categorical variable, and plot them six different ways

menavegarerate = goodreads.loc[(goodreads['author_gender']=='male'),'author_average_rating']
womenavegarerate = goodreads.loc[(goodreads['author_gender']=='female'),'author_average_rating']

plt.hist(menavegarerate,color='orange',alpha=.5,label='Male Author Avg Rate')
plt.hist(womenavegarerate,color='blue',alpha=.5,label='Female Author Avg Rate')
plt.xlabel('Average Rating')
plt.legend(loc='upper left')
plt.title('Plot 1: Average Rating per Gender ')
plt.show()

sns.countplot(y="author_gender", data=goodreads)
sns.catplot(x="author_gender", y="author_average_rating", kind="boxen",data=goodreads)
sns.catplot(x="author_gender", y="author_average_rating", data=goodreads)
sns.catplot(x="author_gender", y="author_average_rating", kind="box", hue='author_gender', data=goodreads)
sns.catplot(x="author_gender",y="author_average_rating", hue="author_gender", kind="violin", data=goodreads)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

#Dataset: Graduate Admissions with parameters which are considered important during the application for Masters Programs. 
#From https://www.kaggle.com/mohansacharya/graduate-admissions/downloads/graduate-admissions.zip/2

admissions = pd.read_csv('Admission_Predict.csv')

#Changing column names
for col in admissions.columns:
    admissions.rename(columns={col:col.replace(" ", "_")},inplace=True)

admissions.rename(columns={'Chance_of_Admit_':'Chance_of_Admit'},inplace=True)
admissions.describe()
```

```python
admissions[(admissions['Chance_of_Admit'] > 0.70) & (admissions['University_Rating'] == 1)]
```

```python
#Creating the master plot

print('Relationships Between different Tests Score and The Chance to Be Admitted in College')
sns.set(style="whitegrid")  
g = sns.pairplot(admissions, 
                  x_vars=["Chance_of_Admit","GRE_Score", "CGPA","TOEFL_Score"],
                  y_vars=["Chance_of_Admit","GRE_Score", "CGPA","TOEFL_Score"],  
                  hue="University_Rating",
                  height=4, palette="husl")
```

## What we are seeing?

We are seeing is a master plot representing some variables which are considered important during the application for Masters Programs. 

The list of variables from left to right:

- Chance of admit (ranging from 0 to 1)
- GRE Score (ranging from 290 to 350)
- Undergraduate GPA Score  (ranging from 6 to 10)
- TOEFL Score  (ranging from 90 to 120)
- and, University Rating  (ranging from 1 to 5)

### Findings:

- The chance to be admitted to University is definitely related with all different test scores.

- The lower test scores are related to the lower ranked Universities

- The higher test scores, the higher chance of admit 

- The higher ranked Universities appears on the right top in all visualizations (violet color), it means the best test scores are very related to highest chances to be admitted to the best Universities
